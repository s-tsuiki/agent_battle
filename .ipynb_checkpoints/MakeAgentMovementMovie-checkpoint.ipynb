```python
#エージェントの動きを確認するためのプログラム
#黒は障害物(壁)、青はプレイヤー1、緑はプレイヤー2、攻撃は赤
#結果はresult/ex_number/date内に出力
from PIL import Image
import numpy as np
import pathlib
import cv2

#各種設定(ここを条件に合わせて変更する)
date = "03-23 13-21-55"  #日付(log/ex2-3(or ex2-4)にある日付のフォルダ名をここにコピペ)
ex_number = "ex2-3"      #課題番号(ex2-3 or ex2-4)
field_width = 5          #フィールドの横の長さ(ここは本体のプログラムをいじらない限り変えなくてよい)
field_height = 5         #フィールドの縦の長さ(ここは本体のプログラムをいじらない限り変えなくてよい)
start_time = 0           #動画と画像を出力する時間の開始時間(この場合、0stepから)
end_time = 1000          #動画と画像を出力する時間の終わり時間(この場合、1000stepまで)default:1000
interval = 1             #動画や画像を出力する間隔(この場合、1stepおき)

#動画用のファイル配列
outimg_files = []
VIDEO_WIDTH = (2 * field_width + 1) * 25
VIDEO_HEIGHT = (2 * field_height + 1) * 25
  
for time in range(start_time, end_time, interval):
        
    #フォルダがない場合は、作成
    p_dir = pathlib.Path("./result/" + ex_number + "/" + date)
    if not p_dir.exists():
        p_dir.mkdir(parents=True)
        
    #指定したファイルに画像を保存
    new_filename = "BattleField_" + str(time) + ".png"
    p_file = p_dir / new_filename
        
    if not p_file.exists():
        filename = "./log/" + ex_number + "/" + date + "/BattleField/battleField_" + str(time) + ".csv"
            
        #filenameのデータを配列に格納
        data = np.loadtxt(filename,
                            dtype = "int",
                            delimiter=",",
                            )
        
        #配列dataのサイズの画像を作成
        height, width = data.shape
        img = Image.new("RGB", (width, height))
        
        #格納されている数字に従って、画像の色を設定
        for y in range(height):
            for x in range(width):
                number = data[y][x]
                #0:赤 1:オレンジ 2:白 3:黒 4:青 5:緑
                if number == 0:
                    img.putpixel((x, y),(255, 0, 0))
                elif number == 1:
                    img.putpixel((x, y),(255, 165, 0))
                elif number == 2:
                    img.putpixel((x, y),(255, 255, 255))
                elif number == 3:
                    img.putpixel((x, y),(0, 0, 0))
                elif number == 4:
                    img.putpixel((x, y),(0, 0, 255))
                elif number == 5:
                    img.putpixel((x, y),(0, 128, 0))
                else:
                    print("({0},{1})でエラー".format(x,y))

        #print("{0},{1}".format(width*5, height*5))
        #そのままだと小さいので、25倍の大きさにする
        img = img.resize(((width*25), (height*25)))
            
        #指定したファイルに画像を保存
        img.save(str(p_dir) + "/" + new_filename)
        
    #画像ファイル追加
    outimg_files.append(str(p_file))

# 動画作成
new_videoname = str(p_dir)+ "/" + "AgentMovement" + str(start_time) + "to" + str(end_time) + ".mp4"
fourcc = cv2.VideoWriter_fourcc('m','p','4', 'v')
video  = cv2.VideoWriter(new_videoname, fourcc, 5.0, (VIDEO_WIDTH, VIDEO_HEIGHT))

for img_file in outimg_files:
    img = cv2.imread(img_file)
    video.write(img)

video.release()
    
#初期化
outimg_files = []

print("Agent Movement Movie released.")
```
